### 2.5 面试题分类与解答

**编程题（LeetCode风格）**:


```python
"""
高频AI岗位编程题
"""

# 1. 反向传播实现
def backprop_cross_entropy(logits, labels):
    """
    实现交叉熵损失的 backward
    logits: (batch, num_classes)
    labels: (batch,) 整数类别
    """
    batch_size = logits.shape[0]
    
    # Softmax
    probs = softmax(logits, axis=-1)
    
    # 梯度 = softmax - one_hot
    grad = probs
    grad[range(batch_size), labels] -= 1
    grad = grad / batch_size
    
    return grad

# 2. 矩阵运算
def matrix_multiply(A, B):
    """实现矩阵乘法"""
    m, n = len(A), len(A[0])
    n2, p = len(B), len(B[0])
    assert n == n2
    
    result = [[0] * p for _ in range(m)]
    for i in range(m):
        for j in range(p):
            for k in range(n):
                result[i][j] += A[i][k] * B[k][j]
    return result

# 3. 梯度检查
def gradient_check(f, x, epsilon=1e-7):
    """数值梯度检查"""
    grad_numeric = []
    grad_analytic = f.backward()
    
    for i in range(len(x)):
        x_plus = x.copy()
        x_plus[i] += epsilon
        x_minus = x.copy()
        x_minus[i] -= epsilon
        
        grad = (f(x_plus) - f(x_minus)) / (2 * epsilon)
        grad_numeric.append(grad)
    
    return np.allclose(grad_numeric, grad_analytic, atol=1e-5)
```

## 3. 代码示例


```python
"""
L24-interview-prep.py
AI算法工程师面试准备工具
"""

import random
from typing import List, Dict

class InterviewQuestionBank:
    def __init__(self):
        self.questions = self.load_questions()
    
    def load_questions(self):
        return {
            "ml_basics": [
                {
                    "question": "解释偏差-方差分解并说明如何处理过拟合",
                    "answer": """
                    偏差-方差分解：
                    - 偏差：模型预测值与真实值的差距，反映模型拟合能力
                    - 方差：模型对训练数据的敏感度，反映模型稳定性
                    - 噪声：数据本身的随机性，无法消除
                    
                    过拟合处理：
                    1. 数据：数据增强、更多数据、数据清洗
                    2. 模型：降低复杂度、L1/L2正则化、Dropout
                    3. 训练：早停、Batch Normalization
                    4. 验证：正确划分数据集、使用交叉验证
                    """,
                    "difficulty": "medium",
                    "frequency": "high"
                },
                {
                    "question": "梯度下降为什么需要学习率？如何选择？",
                    "answer": """
                    学习率作用：控制参数更新的步长
                    - 学习率过大：震荡不收敛，可能错过最优解
                    - 学习率过小：收敛慢，容易陷入局部最优
                    
                    选择策略：
                    1. 经验：常用值 0.001, 0.01, 0.1
                    2. 网格搜索：尝试多个值
                    3. 自适应：Adam等优化器自动调整
                    4. 学习率调度：Warmup + CosineAnnealing
                    """,
                    "difficulty": "easy",
                    "frequency": "high"
                }
            ],
            "deep_learning": [
                {
                    "question": "解释Transformer中的自注意力机制",
                    "answer": """
                    自注意力机制：
                    1. 将输入分为Q（查询）、K（键）、V（值）三个矩阵
                    2. 计算Q和K的点积得到注意力分数
                    3. 除以sqrt(d_k)进行缩放
                    4. softmax归一化得到注意力权重
                    5. 注意力权重与V相乘得到加权输出
                    
                    多头注意力：多个注意力头并行，捕捉不同类型的依赖关系
                    """,
                    "difficulty": "hard",
                    "frequency": "very_high"
                }
            ],
            "coding": [
                {
                    "question": "实现一个简单的神经网络前向传播",
                    "answer": """
                    import numpy as np
                    
                    class SimpleNN:
                        def __init__(self, input_size, hidden_size, output_size):
                            self.W1 = np.random.randn(input_size, hidden_size) * 0.01
                            self.b1 = np.zeros((1, hidden_size))
                            self.W2 = np.random.randn(hidden_size, output_size) * 0.01
                            self.b2 = np.zeros((1, output_size))
                        
                        def relu(self, x):
                            return np.maximum(0, x)
                        
                        def forward(self, X):
                            self.z1 = X @ self.W1 + self.b1
                            self.a1 = self.relu(self.z1)
                            self.z2 = self.a1 @ self.W2 + self.b2
                            return self.z2
                    """,
                    "difficulty": "medium",
                    "frequency": "high"
                }
            ]
        }
    
    def get_random_questions(self, topic, count=5):
        """获取随机面试题"""
        questions = self.questions.get(topic, [])
        return random.sample(questions, min(count, len(questions)))
    
    def simulate_interview(self, duration_minutes=30):
        """模拟面试"""
        topics = list(self.questions.keys())
        print(f"模拟{duration_minutes}分钟面试...")
        print(f"题目类型：{topics}")
        print("\n提示：实际面试时需要口头回答，时间控制很重要")

# 使用示例
bank = InterviewQuestionBank()

# 获取随机题目
ml_questions = bank.get_random_questions("ml_basics", count=3)
print("机器学习基础面试题：")
for i, q in enumerate(ml_questions, 1):
    print(f"\n{i}. {q['question']}")
    print(f"   难度：{q['difficulty']} | 频率：{q['frequency']}")
```
